```python
import os
import logging
import requests
import io
```

```python
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

```python
class SmartShareClient:
    BASE_URL = "https://www.selbetti.com.br/smartshare/SmartShareAPI"

    def __init__(self,
                 usuario: str,
                 senha: str,
                 ds_cliente: str,
                 ds_chave: str) -> None:
        self.session = requests.Session()
        # cabeçalhos genéricos (sem Content-Type!)
        self.session.headers.update({
            "dsCliente":           ds_cliente,
            "dsChaveAutenticacao": ds_chave,
        })
        self.usuario = usuario
        self.senha = senha
        self.token = None
```

```python
def anexar_comprovante(self,
                           cd_fluxo: int,
                           cd_tarefa: int,
                           cd_tipo_anexo: int,#Código no Share do tipo do arquivo ex arquivo Cotação
                           ds_anexo: str,   #Descrição dada pelo usuário
                           file_field: str, #nome do arquivo
                           file_path: str) -> None: #Caminho do arquivo
        url = f"{self.BASE_URL}/api/v1/Anexo/InserirAnexo"
        headers = {
            # Postman headers
            "User-Agent":            "PostmanRuntime/7.44.1",
            "Accept":                "*/*",
            "Accept-Encoding":       "gzip, deflate, br",
            "Connection":            "keep-alive",
            "dsCliente":             self.session.headers["dsCliente"],
            "dsChaveAutenticacao":   self.session.headers["dsChaveAutenticacao"],
            "tokenUsuario":          self.token,
            "cdFluxo":               str(cd_fluxo),
            "cdTarefa":              str(cd_tarefa),
            "cdTipoAnexo":           str(cd_tipo_anexo),
            "dsAnexo":               ds_anexo,
            "dsNomeArquivoOriginal": os.path.basename(file_path),
        }
        if not os.path.isfile(file_path):
            raise FileNotFoundError(f"Arquivo não encontrado: {file_path}")
        with open(file_path, "rb") as f:
            files = {file_field: (os.path.basename(file_path), f, "application/pdf")}
            resp = self.session.post(url, headers=headers, files=files, timeout=60)
        resp.raise_for_status()
        logger.info("Anexo enviado com sucesso.")
```
